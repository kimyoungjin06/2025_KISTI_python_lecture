```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import string
from matplotlib.ticker import LogLocator, NullFormatter


# Parameters
N_ROW = 1
N_COL = 1
X_SIZE = 3.4
Y_SIZE = 3
DPI = 300
plt.rcParams['font.family'] = ['Helvetica', 'NanumSquare']

# Color palette
palette = sns.color_palette("Set2")
linewidth = .8
ahpha=1.
markersize = 2.3


# Line styles & markers
line_styles = ['-', '--', ':']
markers = ['o', 's', '^']
labels = ['actor', 'repo', 'org']
bot_label = {True: "-bot", False: ""}

# Figure setup
fig = plt.figure(figsize=(X_SIZE*N_COL, Y_SIZE*N_ROW), dpi=DPI, facecolor="none")
spec = gridspec.GridSpec(ncols=N_COL, nrows=N_ROW, figure=fig)
axi = 0
ax = fig.add_subplot(spec[axi//N_COL, axi%N_COL])

# Subplot label
ax.text(-.08, 1.05, f'{string.ascii_uppercase[axi]}', transform=ax.transAxes,
        size=12, weight='bold')

# Plot lines with distinct styles
for idx, bot in enumerate([False, True]):
    for j, data in enumerate([pull_cnt_actor, pull_cnt_repo, pull_cnt_org]):
        # 기본 스타일 설정
        ls = line_styles[idx]
        label = f"{labels[j]}{bot_label[bot]}"
        series = data.xs(bot).reset_index(drop=True)

        N = len(series)
        # 로그 스케일 대응: 지수 간격으로 인덱스 선택 (예: 10개 정도만)
        log_indices = np.unique(np.logspace(0, np.log10(N - 1), num=20, dtype=int))

        # 조건부 dashes
        if ls == '--':
            ax.plot(series, label=label,
                    color=palette[j], linestyle=ls, linewidth=linewidth, alpha=ahpha*.66, dashes=(6, 3), 
                   marker=markers[idx], markersize=markersize, markevery=log_indices, markeredgecolor="none")
        else:
            ax.plot(series, label=label,
                    color=palette[j], linestyle=ls, linewidth=linewidth, alpha=ahpha, 
                    marker=markers[idx], markersize=markersize, markevery=log_indices, markeredgecolor="none")
            
# Axes settings
ax.set_xscale("log")
ax.set_yscale("log")
ax.grid(True, which='both', linestyle='--', linewidth=0.12, alpha=0.7)

# Labels
ax.set_xlabel("Rank", fontsize=10)
ax.set_ylabel("Count", fontsize=10)

# Legend
legend = ax.legend(fontsize=4, loc='upper right', frameon=False, ncol=2, handlelength=2.4)
legend.set_title('')

# X축: 메이저 & 마이너 설정
ax.xaxis.set_major_locator(LogLocator(base=10.0, subs=(1.0,), numticks=10))
ax.xaxis.set_minor_locator(LogLocator(base=10.0, subs='auto', numticks=100))
plt.xlim([.9, 1e5])
plt.ylim([.9, 3e4])
# 만약 마이너 틱 라벨 숨기고 싶으면:
ax.xaxis.set_minor_formatter(NullFormatter())

# 원하는 xticks 위치 & 라벨 지정 (선택사항)
ax.set_xticks([1e0, 1e1, 1e2, 1e3, 1e4, 1e5])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import string

# Plot settings
N_ORG = 8
N_ROW = 2
N_COL = 4
X_SIZE = 2.3
Y_SIZE = 2.3
DPI = 300
plt.rcParams['font.family'] = ['Helvetica', 'NanumSquare']

# Data
org_count = df_unq.repo__name.apply(lambda x: x.split("/")[0]).value_counts()
target_orgs = org_count.iloc[:N_ORG].index

# Stars/Forks 전체 (배경용)
stars_all = stars
forks_all = forks

# Plot setup
fig = plt.figure(figsize=(X_SIZE * N_COL, Y_SIZE * N_ROW), dpi=DPI, facecolor="none")
spec = gridspec.GridSpec(ncols=N_COL, nrows=N_ROW, figure=fig)
palette = sns.color_palette("Set2", n_colors=N_ORG)

for i, org in enumerate(target_orgs):
    ax = fig.add_subplot(spec[i // N_COL, i % N_COL], facecolor="none") #, facecolor="none"
    
    # Subplot label (A, B, C...)
    ax.text(-.05, 1.05, f'{string.ascii_uppercase[i]}', transform=ax.transAxes,
            size=12, weight='bold')
    
    # 전체 배경 scatter (회색)
    ax.scatter(stars_all, forks_all, zorder=1, alpha=0.2, s=8, edgecolor="none", color='lightgrey', label='Total')

    # 해당 조직만 필터링
    msk = df_unq['repo__name'].str.contains(org)
    ax.scatter(stars[msk], forks[msk], zorder=2, alpha=0.6, s=10,
               edgecolor="none", color=palette[i], label=org)
    
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_title(org, fontsize=9)
    ax.grid(True, which="both", linestyle='--', linewidth=.3, alpha=.7, zorder=0)
    
    # X/Y축 레이블은 맨 왼쪽/아래만
    if i % N_COL == 0:
        ax.set_ylabel("Forks", fontsize=9)
    else:
        ax.set_yticklabels([])
    if i // N_COL == N_ROW - 1:
        ax.set_xlabel("Stars", fontsize=9)
    else:
        ax.set_xticklabels([])
    
    ax.tick_params(labelsize=7)

plt.tight_layout()
plt.show()
```
