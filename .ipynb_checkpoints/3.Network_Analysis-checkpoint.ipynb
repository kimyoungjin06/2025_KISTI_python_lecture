```python
# Create a bipartite matrix: rows=PR_base_language, columns=repo__name
bipartite_matrix = pd.crosstab(df_g['PR_base_language'], df_g['actor__name'])

print(bipartite_matrix.shape)  # (languages, repos)
```

```python
# í–‰ë ¬ ê³± (languages Ã— repos) * (repos Ã— languages).T â†’ (languages Ã— languages)
proj_matrix = bipartite_matrix @ bipartite_matrix.T

# ìžê¸° ìžì‹ ê³¼ì˜ ì—°ê²° ì œì™¸ (diagonal 0)
np.fill_diagonal(proj_matrix.values, 0)

print(proj_matrix)
```

```python
from scipy.sparse import csr_matrix

# Mapping
lang_idx = {lang: idx for idx, lang in enumerate(df_g['PR_base_language'].unique())}
repo_idx = {repo: idx for idx, repo in enumerate(df_g['actor__name'].unique())}

rows = df_g['PR_base_language'].map(lang_idx)
cols = df_g['actor__name'].map(repo_idx)

# Sparse incidence matrix
data = np.ones(len(df_g))
incidence = csr_matrix((data, (rows, cols)), shape=(len(lang_idx), len(repo_idx)))

# Projection
proj = incidence @ incidence.T
proj.setdiag(0)  # remove self-loops
```

```python
from scipy.sparse import csr_matrix

# proj: csr_matrix í˜•íƒœ (languages Ã— languages)

# 1. coo formatìœ¼ë¡œ ë³€í™˜ (row, col, data ì ‘ê·¼ ì‰½ê²Œ)
proj_coo = proj.tocoo()

# 2. row, col, weight ì¶”ì¶œ
rows = proj_coo.row
cols = proj_coo.col
weights = proj_coo.data

# 3. ìžê¸°ìžì‹  ì—°ê²° ì œê±°
mask = rows != cols
rows = rows[mask]
cols = cols[mask]
weights = weights[mask]

# 4. language ì´ë¦„ ë³µì›
lang_list = list(lang_idx.keys())  # earlier created mapping

edgelist = pd.DataFrame({
    'lang1': [lang_list[i] for i in rows],
    'lang2': [lang_list[j] for j in cols],
    'weight': weights
})

print(edgelist.head())
```

```python
# ë¹ˆ ê·¸ëž˜í”„ ìƒì„±
G = nx.Graph()

# ì—£ì§€ ì¶”ê°€ (weight í¬í•¨)
for row in edgelist.itertuples(index=False):
    G.add_edge(row.lang1, row.lang2, weight=row.weight)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
import matplotlib.pyplot as plt

# í¬ì§€ì…˜ ìžë™ ë°°ì¹˜
pos = nx.spring_layout(G, seed=42)

# ì—£ì§€ ê°€ì¤‘ì¹˜ ì¶”ì¶œ
weights = [G[u][v]['weight'] for u, v in G.edges()]

# ë…¸ë“œ & ì—£ì§€ ê·¸ë¦¬ê¸°
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color='royalblue', node_size=500)
nx.draw_networkx_edges(G, pos, width=[w * 0.0000001 for w in weights], alpha=0.6)
nx.draw_networkx_labels(G, pos, font_size=10)
plt.axis('off')
plt.title("Language Co-occurrence Network")
plt.show()
```

```python
# Specify the initial list of nodes
cate_code_sub = ['Natural'] # , 'Ingredient/substance'
cate_code_neighbor = ['Efficacy', 'Mechanism', 'Ingredient/substance']
node_list = set()
nodelist_nei = set()
for cate in cate_code_sub:
    node_list.update(node_cates[cate])
for cate in cate_code_neighbor:
    nodelist_nei.update(node_cates[cate])


cate_code_sub_tot = cate_code_neighbor + cate_code_sub
colors = ['C0', 'C1', 'C2', 'C3']

# Set the value of k (number of hops/neighbors to consider)
k = 1

# Initialize the set with the original nodes
k_hop_neighbors = set(nodelist)

# Iteratively expand the set by adding neighbors
for _ in range(k):
    # Find all neighbors of the current set of nodes
    neighbors = set()
    for node in k_hop_neighbors:
        neighbors.update(G.neighbors(node))
    
    # Update the set of k-hop neighbors
    k_hop_neighbors.update(neighbors & nodelist_nei)
node_tot = k_hop_neighbors | node_list# | nodelist_nei
# Create the subgraph containing these nodes and their connections
G_sub = G.subgraph(node_tot).copy()

mst = nx.algorithms.tree.maximum_spanning_tree(G_sub)
```
