```python
df = pd.read_parquet("sample.parquet")

df
```

```python
df_unq = df.drop_duplicates(subset=['actor__name', 'repo__name'], keep="last")
len(df_unq)
```

```python
# repo별 PR 수 집계
repo_counts = df_unq['repo__name'].value_counts()

# 10개 초과인 저장소만 선택
popular_repos = repo_counts[repo_counts > 10].index

# 해당 저장소만 남기기
df_masked = df_unq[df_unq['repo__name'].isin(popular_repos)]
```

```python
df_masked
```

```python
repo_counts.iloc[:40]
```

```python
import networkx as nx

data = df_masked[["actor__name", "repo__name"]].itertuples(index=False, name=None)

G = nx.Graph()
G.add_edges_from(data)  # (user, repo)
```

```python
!pip install node2vec
```

```python
from node2vec import Node2Vec
from gensim.models import Word2Vec

# Node2Vec 모델 생성
node2vec = Node2Vec(G, dimensions=128, walk_length=20, num_walks=10, workers=2)

# Random walks를 Gensim Word2Vec 학습에 활용
walks = node2vec.walks  # 또는: node2vec.fit().corpus
walks_str = [[str(node) for node in walk] for walk in walks]
model = Word2Vec(walks_str, vector_size=128, window=5, sg=1, min_count=1)
```

```python
# 유저 벡터
# vec_user = model.wv['some_username']

# 유사한 저장소
model.wv.most_similar('pytorch/pytorch', topn=5)
```

```python
sims = [a for a,s in model.wv.most_similar('pytorch/pytorch', topn=30)]
```

```python
user_cnt = df.actor__name.value_counts()

nodes = [n for n in G.nodes() if '/' not in n and user_cnt.get(n, 0) > 40] # 조건문 순서 유지 필수
nodes += [n for n in G.nodes() if '/' in n]
len(nodes)
```

```python
len([n for n in G.nodes() if '/' not in n and user_cnt.get(n, 0) > 40])
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# 유저 or 저장소만 시각화
embeddings = [model.wv[n] for n in nodes]
embeddings = np.array(embeddings)

tsne = TSNE(n_components=2, random_state=42, perplexity=200)
X_2d = tsne.fit_transform(embeddings)

plt.figure(figsize=(6, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], alpha=0.6)
plt.title("Node2Vec Embedding (t-SNE)")
plt.show()
```

with [jina-embeddings-v3](https://huggingface.co/jinaai/jina-embeddings-v3)

```python
from transformers import AutoModel

# Initialize the model
model = AutoModel.from_pretrained("jinaai/jina-embeddings-v3", trust_remote_code=True)

texts = [
    "Follow the white rabbit.",  # English
    "Sigue al conejo blanco.",  # Spanish
    "Suis le lapin blanc.",  # French
    "跟着白兔走。",  # Chinese
    "اتبع الأرنب الأبيض.",  # Arabic
    "Folge dem weißen Kaninchen.",  # German
]

# When calling the `encode` function, you can choose a `task` based on the use case:
# 'retrieval.query', 'retrieval.passage', 'separation', 'classification', 'text-matching'
# Alternatively, you can choose not to pass a `task`, and no specific LoRA adapter will be used.
embeddings = model.encode(texts, task="text-matching")

# Compute similarities
print(embeddings[0] @ embeddings[1].T)
```
